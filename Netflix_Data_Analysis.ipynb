```python
# EXPLORARION SUMMRY
# * dataframe consisting of 9827 rows and 9 cols
# * no NaNs nor duplicated
# * Release_date column need to be change into  date time and to extract only the year value.
# * Overview , Original_Language and Poster-url wouldn't be useful during analysis, so we drop them.
# * there is noticable outliers in popularity column
# * vote_average better be categoried for proper column
# * Genre column has comma saperated values and white spaces that needs to be handled and casted into category.
```

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
print(df['Release_Date'].dtypes)
```

```python
df['Release_Date'] = df['Release_Date'].dt.year
df['Release_Date'].dtypes
```

```python
df.head()
```

```python
# dropping the columns that are not important
cols = ['Overview','Original_Language','Poster_Url']
```

```python
df.drop(cols , axis = 1 , inplace = True)
df.columns
```

```python
df.head()
```

```python
# categorized vote_average columns(popular,average,below_average,not_popular)
def catigorize_col(df,col,labels):

  edges =[df[col].describe()['min'],
          df[col].describe()['25%'],
          df[col].describe()['50%'],
          df[col].describe()['75%'],
          df[col].describe()['max']]

  df[col] = pd.cut(df[col],edges,labels=labels,duplicates='drop')
  return df

labels = ['not_popular','below_avg','average','popular']
catigorize_col(df,'Vote_Average',labels)
df['Vote_Average'].unique()
```

```python
df.head()
```

```python
df['Vote_Average'].value_counts()
```

```python
df.dropna(inplace = True)
df.isna().sum()
```

```python
df.head()
```

```python
# we'd split genres into a list and explode our dataframe to have only one genre per row for ezch movie

df['Genre'] = df['Genre'].str.split(', ')
df = df.explode('Genre').reset_index(drop=True)
df.head()
```

```python
df['Genre'] = df['Genre'].astype('category')
df['Genre'].dtypes
```

```python
df.info()
```

```python
# DATA VISUALIZATION
sns.catplot(y = 'Genre', data = df,kind = 'count',order = df['Genre'].value_counts().index,color = '#4287f4')
plt.title('Genre column distribution')
plt.show()
```

```python
# useing seaborn functions for visualizing

sns.jointplot(x='Popularity', y = 'Vote_Count' , data = df , kind = 'scatter')
plt.show()
```

```python
# which has highest votes in vote avg column ?
sns.catplot(y = 'Vote_Average', data = df, kind = 'count', order = df['Vote_Average'].value_counts().index,color='#4287f5')
plt.title('Vote_Distribution')
plt.show()
```
